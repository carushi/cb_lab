```python
from sklearn.datasets import load_iris, load_breast_cancer, load_diabetes
```

## Tutorial using iris dataset

```python
iris = load_iris()
print(iris.DESCR)
```

```python
X, y = load_iris(return_X_y=True)

print(X) # 説明変数 Explanatory var
         # Matrix (n x p) sepal length, sepal width, petal length, petal width
print(y) # 目的変数 Response var 0='setosa', 1='setosa', 2='versicolor'
         # Vector (n)
```

```python
# 機械学習器をインポートする
# 予測
from sklearn import svm

clf = svm.SVC()
clf.fit(X, y)
clf.predict([[7.5, 3.0, 5.1, 1.8]])
```

```python
from sklearn.linear_model import LogisticRegression

clf = LogisticRegression(random_state=0).fit(X, y)
clf.predict(X[:2, :])
```

```python
clf.predict_proba(X[:2, :])
```

```python
clf.score(X, y)
```

## Cross validation using breast cancer dataset

```python
breast = load_breast_cancer()
print(breast.DESCR)
```

```python
X, y = load_breast_cancer(return_X_y=True)
print(X)
print(y)
```

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score
import numpy as np

cv = StratifiedKFold(n_splits=3, shuffle=True, random_state=0)
accuracies = []

for train_idx, val_idx in cv.split(X, y):
    X_train = X[train_idx,:]
    y_train = y[train_idx]
    X_val   = X[val_idx,:]
    y_val   = y[val_idx]

    clf.fit(X_train, y_train)

    y_pred = clf.predict(X_val)

    acc = accuracy_score(y_val, y_pred)
    accuracies.append(acc)

print("Average Accuracy: ", np.mean(accuracies))
print(accuracies)
```

## Regression using diabetes dataset

```python
diabetes = load_diabetes()
print(diabetes.DESCR)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler
```

```python
X, y = load_diabetes(return_X_y=True)
print(X) # age, sex, bmi, bp, s1, s2, s3, s4, s5, s6
print(y) # 1年後の疾患の進行度合い
```

```python
Random_forest = RandomForestRegressor(n_estimators=100, max_depth=5)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(np.array(X))

Random_forest.fit(X_scaled, y)
```

```python
# Get predictions
y_pred = Random_forest.predict(X_scaled)

# Compute RMSE
print("RMSE:",np.sqrt(mean_squared_error(y,y_pred)))
```
